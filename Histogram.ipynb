```python
import numpy as np # Mengimpor library numpy untuk manipulasi data numerik
import imageio # Mengimpor library imageio untuk membaca gambar
import matplotlib.pyplot as plt # Mengimpor library matplotlib untuk menampilkan gambar

img = imageio.imread("messi.jpg") # Membaca gambar "messi.jpg" menggunakan imageio
img_height = img.shape[0] # Mengambil tinggi gambar dari objek gambar
img_width = img.shape[1] # Mengambil lebar gambar dari objek gambar
img_channel = img.shape[2] # Mengambil jumlah channel dari objek gambar

# Membuat array kosong dengan ukuran dan tipe data yang sama dengan gambar untuk menyimpan gambar grayscale
img_grayscale = np.zeros(img.shape, dtype=np.uint8)

# Looping untuk mengubah setiap piksel menjadi grayscale
for y in range(0, img_height):
    for x in range(0, img_width):
        red = img[y][x][0] # Mengambil nilai merah dari piksel
        green = img[y][x][1] # Mengambil nilai hijau dari piksel
        blue = img[y][x][2] # Mengambil nilai biru dari piksel
        gray = (int(red) + int(green) + int(blue)) / 3 # Menghitung rata-rata dari nilai merah, hijau, dan biru
        img_grayscale[y][x] = (gray, gray, gray) # Menyimpan nilai grayscale ke array yang telah dibuat

# Menampilkan gambar grayscale
plt.imshow(img_grayscale)
plt.title("Grayscale")
plt.show()
```

```python
hg = np.zeros((256)) # Membuat array kosong dengan 256 elemen untuk menyimpan histogram grayscale

for x in range(0, 256): # Looping untuk menginisialisasi semua elemen array hg dengan nilai 0
    hg[x] = 0

# Looping untuk menghitung histogram grayscale
for y in range(0, img_height):
    for x in range(0, img_width):
        gray = img_grayscale[y][x][0] # Mengambil nilai grayscale dari piksel
        hg[gray] += 1 # Menambahkan hitungan untuk nilai grayscale yang sesuai

# Membuat 100 bins untuk plot histogram
bins = np.linspace(0, 256, 100)

# Menampilkan plot histogram grayscale
plt.hist(hg, bins, color="black", alpha=0.5)
plt.title("Histogram")
plt.show()
```

```python
# Inisialisasi array untuk histogram masing-masing warna dan kombinasi RGB
hgr = np.zeros((256))
hgg = np.zeros((256))
hgb = np.zeros((256))
hgrgb = np.zeros((768))

# Looping untuk menginisialisasi semua elemen array histogram dengan nilai 0
for x in range(0, 256):
    hgr[x] = 0
    hgg[x] = 0
    hgb[x] = 0
    
for x in range(0, 768):
    hgrgb[x] = 0
    
# Looping untuk menghitung histogram warna dan kombinasi RGB
for y in range(0, img.shape[0]):
    for x in range(0, img.shape[1]):
        # Mengambil nilai warna dari piksel dan menambahkan 256 dan 512 untuk green dan blue
        red = int(img[y][x][0])
        green = int(img[y][x][1])
        blue = int(img[y][x][2])
        green = green + 256
        blue = blue + 512
        # Menambahkan hitungan untuk nilai warna dan kombinasi RGB yang sesuai
        hgr[red] += 1
        hgg[green-256] += 1
        hgb[blue-512] += 1
        hgrgb[red] += 1
        hgrgb[green] += 1
        hgrgb[blue] += 1

# Membuat 100 bins untuk plot histogram RGB
binsrgb = np.linspace(0, 768, 100)

# Menampilkan plot histogram RGB
plt.hist(hgrgb, binsrgb, color="black", alpha=0.5)
plt.title("Histogram Red Green Blue")
plt.show()
```

```python
hgk = np.zeros((256)) # inisialisasi array hgk dengan nilai nol sebanyak 256 elemen
c = np.zeros((256)) #inisialisasi array c dengan nilai nol sebanyak 256 elemen

#mengisi array hgk dan c dengan nilai nol menggunakan perulangan for
for x in range(0, 256): 
    hgk[x] = 0
    c[x] = 0

#melakukan perulangan for pada setiap piksel gambar, mengambil nilai grayscale dari piksel tersebut, kemudian menambahkan nilai histogram pada array hgk sesuai dengan nilai grayscale tersebut
for y in range(0, img_height): 
    for x in range(0, img_width): 
        gray = img_grayscale[y][x][0] 
        hgk[gray] += 1
                
c[0] = hgk[0] #set nilai awal array c dengan nilai hgk pada indeks 0
#menghitung nilai kumulatif dengan menjumlahkan nilai array hgk dengan nilai sebelumnya pada array c
for x in range(1, 256): 
     c[x] = c[x-1] + hgk[x]

hmaxk = c[255] #set variabel hmaxk dengan nilai terakhir pada array c

#melakukan normalisasi dan scaling pada nilai array c dengan mengalikan setiap nilai dengan faktor 190 dan dibagi dengan nilai hmaxk
for x in range(0, 256):
    c[x] = 190 * c[x] / hmaxk

#menampilkan histogram dari array c dengan menggunakan 100 bin dan warna hitam, kemudian menampilkan judul grafik histogram
plt.hist(c, bins, color="black", alpha=0.5)
plt.title("Histogram Grayscale Kumulatif")
plt.show()
```

```python
hgh = np.zeros((256))
h = np.zeros((256))
c = np.zeros((256))

#inisialisasi array hgh dan h dengan 0
for x in range(0, 256):
    hgh[x] = 0
    h[x] = 0
    c[x] = 0

#melakukan iterasi pada setiap piksel dalam citra grayscale untuk menghitung frekuensi kemunculan nilai piksel
for y in range(0, img_height):
    for x in range(0, img_width):
        gray = img_grayscale[y][x][0]
        hgh[gray] += 1

#menghitung kumulatif frekuensi kemunculan nilai piksel             
h[0] = hgh[0]
for x in range(1, 256):
     h[x] = h[x-1] + hgh[x]

#menghitung probabilitas kemunculan nilai piksel
for x in range(0, 256):
     h[x] = h[x] / img_height / img_width

#mengosongkan array hgh untuk menghitung frekuensi kemunculan nilai piksel setelah histogram equalization
for x in range(0, 256):
    hgh[x] = 0

#melakukan iterasi pada setiap piksel dalam citra grayscale untuk melakukan histogram equalization   
for y in range(0, img_height):
    for x in range(0, img_width):
        gray = img_grayscale[y][x][0]
        gray = h[gray] * 255 #menghitung nilai piksel yang setelah histogram equalization
        hgh[int(gray)] += 1 #menghitung frekuensi kemunculan nilai piksel setelah histogram equalization

#menghitung kumulatif frekuensi kemunculan nilai piksel setelah histogram equalization
c[0] = hgh[0]
for x in range(1, 256):
     c[x] = c[x-1] + hgh[x]

hmaxk = c[255] #mendapatkan nilai maksimum frekuensi kemunculan nilai piksel setelah histogram equalization

#menghitung nilai piksel setelah dilakukan kumulatif histogram equalization
for x in range(0, 256):
    c[x] = 190 * c[x] / hmaxk

#menampilkan histogram setelah dilakukan histogram equalization
plt.hist(c, bins, color="black", alpha=0.5)
plt.title("Histogram Grayscale Hequalisasi")
plt.show()
```
