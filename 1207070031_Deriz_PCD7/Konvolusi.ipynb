```python
#Baris ini mengimpor modul matplotlib.pyplot dan mengaktifkan plotting secara inline di Jupyter Notebook.
import matplotlib.pyplot as plt
%matplotlib inline
#Baris ini mengimpor beberapa modul dari library scikit-image (skimage). Modul data menyediakan contoh gambar, imread digunakan untuk membaca gambar, rgb2gray mengonversi gambar berwarna menjadi grayscale, dan numpy adalah library yang banyak digunakan untuk operasi numerik.
from skimage import data
from skimage.io import imread
from skimage.color import rgb2gray 
import numpy as np
#Baris ini mengimpor library OpenCV, yang menyediakan berbagai fungsi dan algoritma dalam bidang visi komputer.
import cv2
```

```python
citra1 = imread(fname="messi.jpg") #Baris ini membaca gambar dengan nama file "messi.jpg" menggunakan fungsi imread dari modul scikit-image. Gambar tersebut akan disimpan dalam variabel citra1.
print(citra1.shape) #Baris ini mencetak dimensi gambar citra1. citra1.shape memberikan output berupa tuple yang menunjukkan (tinggi, lebar, jumlah saluran warna) dari gambar.

plt.imshow(citra1, cmap='gray') #Baris ini menampilkan gambar citra1 menggunakan fungsi imshow dari modul matplotlib.pyplot. Argumen citra1 digunakan sebagai gambar yang akan ditampilkan, sedangkan cmap='gray' mengatur skema warna menjadi skala abu-abu.
```

```python
#Baris ini mendefinisikan sebuah kernel sebagai matriks numpy dengan ukuran 3x3. Kernel ini digunakan dalam proses filtering atau konvolusi pada gambar.
kernel = np.array([[-1, 0, -1], 
                   [0, 4, 0], 
                   [-1, 0, -1]])
#Baris ini melakukan proses filtering atau konvolusi pada citra1 menggunakan kernel yang telah didefinisikan sebelumnya. Hasil filtering disimpan dalam variabel citraOutput.
citraOutput = cv2.filter2D(citra1, -1, kernel)
#Baris ini menampilkan gambar citra1 dan citraOutput menggunakan imshow dari modul matplotlib.pyplot. Gambar-gambar tersebut ditampilkan dalam satu baris dengan ukuran gambar yang ditentukan oleh figsize. Setiap gambar diberi judul menggunakan set_title.
fig, axes = plt.subplots(1, 2, figsize=(12, 12))
ax = axes.ravel()

ax[0].imshow(citra1, cmap = 'gray')
ax[0].set_title("Citra Input")
ax[1].imshow(citraOutput, cmap = 'gray')
ax[1].set_title("Citra Output")
```
