```python
import matplotlib.pyplot as plt #Baris ini mengimpor library Matplotlib yang digunakan untuk visualisasi data dan plotting.
%matplotlib inline #Baris ini adalah sebuah magic command untuk memastikan bahwa output plot dari Matplotlib akan ditampilkan di dalam notebook Jupyter.

from skimage import data #Baris ini mengimpor modul "data" dari library scikit-image yang menyediakan kumpulan gambar percobaan.
from skimage.io import imread #Baris ini mengimpor fungsi "imread" dari modul "io" dalam library scikit-image. Fungsi ini digunakan untuk membaca gambar dari file.
from skimage.color import rgb2gray #Baris ini mengimpor fungsi "rgb2gray" dari modul "color" dalam library scikit-image. Fungsi ini digunakan untuk mengonversi gambar dari format RGB ke skala abu-abu.
import numpy as np #Baris ini mengimpor library NumPy yang digunakan untuk operasi matematika dan manipulasi array.
import cv2
```

```python
#Baris ini membaca dua file gambar "messi.jpg" dan "ronaldo.jpg" menggunakan fungsi imread dari scikit-image. Gambar tersebut disimpan dalam variabel citra1 dan citra2.Baris ini membaca dua file gambar "messi.jpg" dan "ronaldo.jpg" menggunakan fungsi imread dari scikit-image. Gambar tersebut disimpan dalam variabel citra1 dan citra2.
citra1 = imread(fname="messi.jpg")
citra1 = cv2.cvtColor(citra1,cv2.COLOR_RGB2GRAY)
citra2 = imread(fname="ronaldo.jpg")
citra2 = cv2.cvtColor(citra2,cv2.COLOR_RGB2GRAY)
#Baris ini mencetak dimensi citra pertama (citra1) dan citra kedua (citra2) menggunakan atribut shape. Ini memberikan informasi tentang jumlah baris, jumlah kolom, dan jumlah saluran warna (jika ada) dari setiap citra.
print('Shape citra 1 : ', citra1.shape)
print('Shape citra 2 : ', citra2.shape)
#Baris ini menggunakan fungsi subplots dari Matplotlib untuk membuat tata letak subplot dengan 1 baris dan 2 kolom. Ukuran total plot ditentukan oleh figsize=(10, 10). Variabel axes dan ax digunakan untuk menyimpan referensi ke objek subplot.
fig, axes = plt.subplots(1, 2, figsize=(10, 10))
ax = axes.ravel()
#Baris ini menampilkan citra pertama (citra1) di subplot pertama (ax[0]) dan citra kedua (citra2) di subplot kedua (ax[1]) menggunakan fungsi imshow dari Matplotlib. Parameter cmap='gray' digunakan untuk menampilkan citra dalam skala abu-abu. Judul plot untuk masing-masing citra diatur menggunakan set_title().
ax[0].imshow(citra1, cmap = 'gray')
ax[0].set_title("Citra 1")
ax[1].imshow(citra2, cmap = 'gray')
ax[1].set_title("Citra 2")
```

```python
#Baris ini membuat salinan dari citra pertama (citra1) dan citra kedua (citra2) menggunakan metode .copy() pada objek citra. Kemudian, tipe data citra diubah menjadi float menggunakan metode .astype(float).
copyCitra1 = citra1.copy().astype(float)
copyCitra2 = citra2.copy().astype(float)
#Baris ini mengambil dimensi citra pertama (copyCitra1) menggunakan atribut .shape dan menyimpannya dalam variabel m1 dan n1. Selanjutnya, array kosong dengan dimensi yang sama dengan citra pertama (copyCitra1) dibuat menggunakan fungsi np.empty(). Array ini akan digunakan untuk menyimpan hasil operasi pada citra.
m1,n1 = copyCitra1.shape
output1 = np.empty([m1, n1])
#Baris ini mengambil dimensi citra kedua (copyCitra2) menggunakan atribut .shape dan menyimpannya dalam variabel m2 dan n2. Selanjutnya, array kosong dengan dimensi yang sama dengan citra kedua (copyCitra2) dibuat menggunakan fungsi np.empty(). Array ini akan digunakan untuk menyimpan hasil operasi pada citra.
m2,n2 = copyCitra2.shape
output2 = np.empty([m2, n2])
#Baris ini mencetak bentuk (shape) dari citra pertama yang telah dicopy (copyCitra1) dan bentuk dari array kosong yang telah dibuat (output1). Variabel m1 dan n1 juga dicetak untuk menampilkan jumlah baris dan kolom citra pertama.
print('Shape copy citra 1 : ', copyCitra1.shape)
print('Shape output citra 1 : ', output1.shape)
print('m1 : ',m1)
print('n1 : ',n1)
print()
#Baris ini mencetak bentuk (shape) dari citra kedua yang telah dicopy (copyCitra2) dan bentuk dari array kosong yang telah dibuat (output2). Variabel m2 dan n2 juga dicetak untuk menampilkan jumlah baris dan kolom citra kedua.
print('Shape copy citra 2 : ', copyCitra2.shape)
print('Shape output citra 3 : ', output2.shape)
print('m2 : ',m2)
print('n2 : ',n2)
print()
```

```python
#Baris ini memulai dua loop nested (loop dalam loop) menggunakan range() untuk iterasi pada setiap baris dan kolom citra pertama.
for baris in range(0, m1-1):
    for kolom in range(0, n1-1):
        #Baris ini menginisialisasi variabel a1 dan b1 dengan nilai baris dan kolom saat ini dalam iterasi. Variabel ini akan digunakan untuk mengakses elemen citra pertama dan menyimpan hasil operasi di output.
        a1 = baris
        b1 = kolom
        #Baris ini menjumlahkan elemen tetangga dari elemen saat ini dalam citra pertama (copyCitra1). Elemen tetangga yang dijumlahkan adalah elemen di sekitar elemen saat ini dalam pola 3x3. Hasil penjumlahan disimpan dalam variabel jumlah.
        jumlah = copyCitra1[a1-1, b1-1] + copyCitra1[a1-1, b1] + copyCitra1[a1-1, b1-1] + \
                 copyCitra1[a1, b1-1] + copyCitra1[a1, b1] + copyCitra1[a1, b1+1] + \
                 copyCitra1[a1+1, b1-1] + copyCitra1[a1+1, b1] + copyCitra1[a1+1, b1+1];  
        #Baris ini menghitung rata-rata dari jumlah dengan membaginya dengan 9 (karena terdapat 9 elemen yang dijumlahkan) dan menyimpan hasilnya di output (output1). Ini merupakan operasi perataan (smoothing) menggunakan rata-rata tetangga pada citra pertama.
        output1[a1, b1] = (1/9 * jumlah)
```

```python
#Baris ini memulai dua loop nested (loop dalam loop) menggunakan range() untuk iterasi pada setiap baris dan kolom citra kedua.
for baris1 in range(0, m2-1):
    for kolom1 in range(0, n2-1):
        #Baris ini menginisialisasi variabel a1 dan b1 dengan nilai baris dan kolom saat ini dalam iterasi. Variabel ini akan digunakan untuk mengakses elemen citra kedua dan menyimpan hasil operasi di output.
        a1 = baris1
        b1 = kolom1
        #Baris ini menjumlahkan elemen tetangga dari elemen saat ini dalam citra kedua (copyCitra2). Elemen tetangga yang dijumlahkan adalah elemen di sekitar elemen saat ini dalam pola 3x3. Hasil penjumlahan disimpan dalam variabel jumlah.
        jumlah = copyCitra2[a1-1, b1-1] + copyCitra2[a1-1, b1] + copyCitra2[a1-1, b1-1] + \
                 copyCitra2[a1, b1-1] + copyCitra2[a1, b1] + copyCitra2[a1, b1+1] + \
                 copyCitra2[a1+1, b1-1] + copyCitra2[a1+1, b1] + copyCitra2[a1+1, b1+1];  
        #Baris ini menghitung rata-rata dari jumlah dengan membaginya dengan 9 (karena terdapat 9 elemen yang dijumlahkan) dan menyimpan hasilnya di output (output2). Ini merupakan operasi perataan (smoothing) menggunakan rata-rata tetangga pada citra kedua.
        output2[a1, b1] = (1/9 * jumlah)
```

```python
#Baris ini membuat sebuah figure dan empat axes (subplots) dengan ukuran 2x2 dan ukuran total 10x10. Variabel ax kemudian menggabungkan semua axes menjadi satu dimensi.
fig, axes = plt.subplots(2, 2, figsize=(10, 10))
ax = axes.ravel()
#Baris ini menampilkan citra pertama (citra1) di axes pertama (ax[0]) dengan menggunakan colormap 'gray' dan mengatur judul subplot menjadi "Input Citra 1".
ax[0].imshow(citra1, cmap = 'gray')
ax[0].set_title("Input Citra 1")
#Baris ini menampilkan citra kedua (citra2) di axes kedua (ax[1]) dengan menggunakan colormap 'gray' dan mengatur judul subplot menjadi "Input Citra 2".
ax[1].imshow(citra2, cmap = 'gray')
ax[1].set_title("Input Citra 1")
#Baris ini menampilkan output citra pertama (output1) di axes ketiga (ax[2]) dengan menggunakan colormap 'gray' dan mengatur judul subplot menjadi "Output Citra 1".
ax[2].imshow(output1, cmap = 'gray')
ax[2].set_title("Output Citra 1")
#Baris ini menampilkan output citra kedua (output2) di axes keempat (ax[3]) dengan menggunakan colormap 'gray' dan mengatur judul subplot menjadi "Output Citra 2".
ax[3].imshow(output2, cmap = 'gray')
ax[3].set_title("Output Citra 2")
```
