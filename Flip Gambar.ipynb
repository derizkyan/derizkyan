```python
import numpy as np # Mengimpor library numpy untuk manipulasi data numerik
import imageio # Mengimpor library imageio untuk membaca gambar
import matplotlib.pyplot as plt # Mengimpor library matplotlib untuk menampilkan gambar

img = imageio.imread("messi.jpg") # Membaca gambar "messi.jpg" menggunakan imageio
img_height = img.shape[0] # Mengambil tinggi gambar dari objek gambar
img_width = img.shape[1] # Mengambil lebar gambar dari objek gambar
img_channel = img.shape[2] # Mengambil jumlah channel dari objek gambar
img_type = img.dtype # Mengambil tipe data gambar dari objek gambar

# Membuat array kosong dengan ukuran dan tipe data yang sama dengan gambar untuk menyimpan gambar yang sudah dibalik
img_flip_horizontal = np.zeros(img.shape, img_type)
img_flip_vertical = np.zeros(img.shape, img_type)

# Looping untuk membalik gambar secara horizontal
for y in range(0, img_height):
    for x in range(0, img_width):
        for c in range(0, img_channel):
            img_flip_horizontal[y][x][c] = img[y][img_width-1-x][c]

# Looping untuk membalik gambar secara vertikal
for y in range(0, img_height):
    for x in range(0, img_width):
        for c in range(0, img_channel):
            img_flip_vertical[y][x][c] = img[img_height-1-y][x][c]

# Menampilkan gambar yang sudah dibalik secara horizontal
plt.imshow(img_flip_horizontal)
plt.title("Flip Horizontal")
plt.show()

# Menampilkan gambar yang sudah dibalik secara vertikal
plt.imshow(img_flip_vertical)
plt.title("Flip Vertical")
plt.show()
```
